```python
df1=pd.DataFrame({
    "city":['new york','chicago','orlando','baltimore'],
    "temperature":[21,14,35,32],
})
df1
```

```python
df2=pd.DataFrame({
    "city":['new york','chicago','san francisco'],
    "humidity":[65,68,71],
})
df2
```

```python
df3=pd.merge(df1,df2,on='city') #performend on city column and gives same data(inner join)
df3
```

```python
df3=pd.merge(df1,df2,on='city', how='outer') #performend on city column and gives all data(outer join)
df3
```

```python
df3=pd.merge(df1,df2,on='city', how='left')
#performed on city column and gives all data from left(left join)

display(df1)
display(df2)
print(df3)
```

```python
df3=pd.merge(df1,df2,on='city', how='right')
#performed on city column and gives all data from right and(right join)

display(df1)
display(df2)
print(df3)
```

```python
df3=pd.merge(df1,df2,on='city', how='left',indicator=True) 
#gives additional column and tells from which dataframe it comes from.

display(df1)
display(df2)
print(df3)
```
